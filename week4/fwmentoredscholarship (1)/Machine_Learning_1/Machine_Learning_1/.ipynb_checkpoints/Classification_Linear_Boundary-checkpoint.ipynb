```python
%reset -f

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
theScores = np.random.randint(55,101,(800,2))
theOutcome = np.where(np.sum(theScores,1)<=155, 0, 1)

df = pd.DataFrame(theScores, columns = ['Test 1', 'Test 2'])

df.insert(2, 'Total', np.sum(theScores, 1))
df.insert(3, 'Outcome', theOutcome)
```

```python
df0 = df[df['Outcome']==0]
df1 = df[df['Outcome']==1]

plt.figure(figsize=(12,8))
plt.plot(df0['Test 1'], df0['Test 2'], 'ro')
plt.plot(df1['Test 1'], df1['Test 2'], 'o')
plt.xlabel('Test 1', fontsize=20)
plt.xticks(fontsize=15)
plt.xlim(54,101)
plt.ylabel('Test 2', fontsize=20)
plt.yticks(fontsize=15)
plt.ylim(54,101)
plt.legend(['Not Admitted', 'Admitted'], loc=1, fontsize=20, markerscale=1.5, frameon=True, bbox_to_anchor=(1.33, 1))
plt.show()
```

```python
theModel = LogisticRegression()
theModel.fit(df[['Test 1', 'Test 2']], df['Outcome']);

print 'Training Set Score: ', 100.0*sum(theModel.predict(df[['Test 1', 'Test 2']]) == theOutcome)/np.shape(theOutcome)[0], '%'
```

```python
# NOTE: h = sigmoid(z)"
#       where z = m0 + m1x1 +m2x2 + ... + mnxn
#       Decision boundary occurs where h=0.5 (i.e. where z=0)

t1, t2 = theModel.coef_[0][0], theModel.coef_[0][1]
x = [55, 100]
y = -1.0*(theModel.intercept_ + np.dot(t1,x))/t2

plt.figure(figsize=(12,8))
plt.plot(df0['Test 1'], df0['Test 2'], 'ro')
plt.plot(df1['Test 1'], df1['Test 2'], 'o')
plt.plot(x, y, 'k-', linewidth=3)
plt.xlabel('Test 1', fontsize=20)
plt.xticks(fontsize=15)
plt.xlim(54,101)
plt.ylabel('Test 2', fontsize=20)
plt.yticks(fontsize=15)
plt.ylim(54,101)
plt.legend(['Not Admitted', 'Admitted'], loc=1, fontsize=20, markerscale=1.5, frameon=True, bbox_to_anchor=(1.33, 1))
plt.show()
```
