```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')

import pandas as pd
import numpy as np

from collections import Counter

import psutil

from colorama import Style, Fore, Back
```

```python
df = pd.DataFrame()
```

NOTE: 'Ft_PS' represents polling station data which should be categorical but is shown here as numerical values.
<br>
These should be converted to categorical values as shown below

```python
n = 10

df['Ft_Num'] = np.random.uniform(-5,5, [n,])
df['Ft_PS'] = np.random.randint(1,10, [n,])
df['Ft_Miles'] = np.random.randint(1,10, [n,])
df['Ft_Cat'] = np.random.choice(['V1', 'V2', 'V3', 'V4'], size=n)

df
```

```python
df['Ft_PS'] = ['PS'+str(i) for i in df['Ft_PS']]
df
```

```python
df = pd.get_dummies(df)
df
```
